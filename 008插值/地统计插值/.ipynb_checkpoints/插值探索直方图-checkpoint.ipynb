```python
import arcpy,pandas
import matplotlib.pyplot as plt
```

```python
def getColor(colName, colNum):
    cm = plt.cm.get_cmap(colName)
    col = [cm(float(i)/(colNum)) for i in range(colNum)]
    cols = []
    for c in col:
        f = "#"
        for c1 in c:
            cc = "{0}".format(hex(int(c1*255)))[2:]
            if len(cc) < 2:
                cc = "0"+cc
            f+=cc
        cols.append(f)
    return cols
```

```python
dt = arcpy.da.FeatureClassToNumPyArray(r"E:\work\地统计插值\weather.shp",
                                       ["mean","Shape@XY"])
```
