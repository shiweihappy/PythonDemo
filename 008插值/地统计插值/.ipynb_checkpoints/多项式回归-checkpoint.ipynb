```python
import arcpy,pandas
import matplotlib.pyplot as plt
import numpy
import scipy.stats as st
```

```python
dt = arcpy.da.FeatureClassToNumPyArray("./data/weather.shp",
                                       ["mean","SHAPE@X","SHAPE@Y"])
```

```python
from sklearn import  linear_model
from sklearn.preprocessing import  PolynomialFeatures
```

```python
def Coff(x,y):
    minX =min(x)
    maxX =max(x)
    X=numpy.arange(minX,maxX).reshape([-1,1])
    poly_reg =PolynomialFeatures(degree=2)
    dx = x.reshape([len(x),1])
    X_ploy =poly_reg.fit_transform(dx)
    lin_reg_2=linear_model.LinearRegression()
    lin_reg_2.fit(X_ploy,y)
    plt.scatter(dx,y,color='blue')
    plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color='red')
    plt.show()
```

```python
Coff(dt["SHAPE@Y"],dt["mean"])
```

```python
Coff(dt["SHAPE@X"],dt["mean"])
```
