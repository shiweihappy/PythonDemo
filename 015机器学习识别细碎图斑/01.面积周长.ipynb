```python
import arcpy,numpy,pandas
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
pandas.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
shp = arcpy.da.FeatureClassToNumPyArray("e:/work/机器学习/shp/jbnt.shp",
                                        ["DLBM","SHAPE_Area","SHAPE_Leng"])
```

```python
pd = pandas.DataFrame(shp)
```

```python
pd.describe()
```

```python
plt.figure(figsize=(12,5))
plt.title("周长分布（频数）")
plt.hist(shp["SHAPE_Leng"],200)
pass
```

```python
plt.figure(figsize=(12,5))
plt.title("周长分布（频率）")
plt.hist(shp["SHAPE_Leng"],200,density=True)
pass
```

```python
plt.figure(figsize=(12,5))
plt.title("周长分布（5000米以内）")
plt.hist(shp["SHAPE_Leng"],200,density=True)
plt.xlim(0,5000)
pass
```

```python
plt.figure(figsize=(12,5))
plt.title("周长分布（2000以内）")
plt.hist(shp["SHAPE_Leng"][shp["SHAPE_Leng"] < 2000],200)
pass
```

```python
plt.figure(figsize=(12,5))
plt.title("面积分布")
plt.hist(shp["SHAPE_Area"],200)
pass
```

```python
plt.figure(figsize=(12,5))
plt.title("面积分布（30万平米以内）")
plt.hist(shp["SHAPE_Area"],200)
plt.xlim(0,300000)
pass
```

```python
plt.figure(figsize=(12,5))
plt.title("面积分布（6万平米以内）")
plt.hist(shp["SHAPE_Area"][shp["SHAPE_Area"] < 60000],100)
pass
```

```python
end = int(len(shp["SHAPE_Area"])*0.9)
```

```python
plt.figure(figsize=(12,5))
plt.title("周长与面积分布（90%以内）")

area90 = numpy.sort(shp["SHAPE_Area"])[:end]
leng90 = numpy.sort(shp["SHAPE_Leng"])[:end]
plt.hist2d(area90,leng90,20)
pass
```
