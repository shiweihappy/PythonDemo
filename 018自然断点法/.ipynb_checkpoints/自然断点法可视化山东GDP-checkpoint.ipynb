```python
import arcpy,pandas,json,jenkspy
```

```python
import seaborn as sns
sns.set(font_scale=1.4)
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def getColorDist(colName, colValue):
    cm = plt.cm.get_cmap(colName)
    colNum = len(colValue)
    col = [cm(float(i)/(colNum)) for i in range(colNum)]
    cols = {}
    for i in range(len(col)):
        c = col[i]
        f = "#"
        for c1 in c:
            cc = "{0}".format(hex(int(c1*255)))[2:]
            if len(cc) < 2:
                cc = "0"+cc
            f+=cc
        cols[colValue[i]] = f
    return cols
```

```python
feat = "../013山东经济数据聚类/data2_SD/shandongProject.shp"
```

```python
sd = pandas.DataFrame(arcpy.da.FeatureClassToNumPyArray(feat,["FIRST_NAME","F"]))
sd.columns=["name","财政收入"]
```

## 设定分为两级

```python
nb_class = 2
```

```python
breaks = jenkspy.jenks_breaks(sd["财政收入"],nb_class = nb_class)
```

## 两级的自然断点如下

```python
print(breaks)
```

```python
def getBreaksNum(breaks,val):
    for i in range(len(breaks)-1):
        if val >= breaks[i] and val <= breaks[i+1]:
            return i
```

```python
col = getColorDist("Set1",range(nb_class))
ext = arcpy.Describe(feat).extent
fig = plt.figure(figsize=(14,9)) 
ax = fig.gca() 
ax.grid(alpha= 1)
basemap = arcpy.FeatureSet(feat)
for g in arcpy.da.SearchCursor(basemap,["SHAPE@JSON","FIRST_NAME",
                                       "SHAPE@XY","F"]):
    clf = getBreaksNum(breaks,g[3])
    c = col[clf]
    poly = plt.Polygon(json.loads(g[0])["rings"][0],
                       ec = '#000000',fc=c, alpha = 0.5)
    plt.text(g[2][0], g[2][1],g[1],fontsize=8)
    ax.add_patch(poly)
ax.set_xlim(ext.XMin,ext.XMax)
ax.set_ylim(ext.YMin,ext.YMax)
plt.show()
```
