```python
import arcpy
```

```python
arcpy.env.workspace = "./shp"
```

```python
for f in arcpy.ListFields("./shp/CNPG_S.shp"):
    print(f.name)
```

```python
with arcpy.da.SearchCursor("CNPG_S.shp",["FIRST_NAME",
                                         "GDP_2009","Pop_2009"]) as cursor:
    for row in cursor:
        print(row)
```
